### Định dạng dữ liệu Thời gian về Date Time


```python
df_aapl["Date"] = pd.to_datetime(df_aapl["Date"])
```

### Kiểm tra dữ liệu thiếu, lặp

```python
missing_value = df_aapl.isnull().sum()
print("Tổng số dữ liệu thiếu:\n", missing_value)
```

```python
duplicate = df_aapl.duplicated().sum()
print(f"Số bản ghi trùng lặp: {duplicate}")
```

### Tính lợi nhuận Logarit

```python
df_aapl['Log_Return'] = np.log(df_aapl['Close'] / df_aapl['Close'].shift(1))
```

### Tính Volatility

```python
df_aapl['Volatility_20d'] = (
    df_aapl['Log_Return']
    .rolling(20)
    .std()
    * np.sqrt(252)
)

df_aapl['Volatility_60d'] = (
    df_aapl['Log_Return']
    .rolling(60)
    .std()
    * np.sqrt(252)
)
```

### Đầu ra

```python
df_aapl.tail()
```
